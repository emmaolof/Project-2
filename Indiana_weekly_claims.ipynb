```python
import pandas as pd
import csv
```

```python
def allfridays(year):
    return pd.date_range(start=str(year), end=str(year+1), 
                         freq='W-FRI').strftime('%Y%m%d').tolist()
```

```python
Indiana_weekly_claims = pd.read_csv("Indiana_Unemployment_Claims.csv")
```

```python
Indiana_weekly_claims.columns = ['County','StateFips','CountyFips','Year', 'Month', 'MonthCode','Week', 'InitialClaims', 'ChangeFrom2019']
```

```python
Indiana_weekly_claims.dtypes
```

```python
Indiana_weekly_claims['CountyFips'] = Indiana_weekly_claims['CountyFips'].astype(str)
Indiana_weekly_claims['CountyFips'] = Indiana_weekly_claims['CountyFips'].apply(lambda x: x.zfill(3))
Indiana_weekly_claims['Fips'] = Indiana_weekly_claims['StateFips'].astype(str) + Indiana_weekly_claims['CountyFips']
Indiana_weekly_claims.fillna(0, inplace=True)
Indiana_weekly_claims['InitialClaims'] = Indiana_weekly_claims['InitialClaims'].astype(str)
Indiana_weekly_claims['InitialClaims'] = Indiana_weekly_claims['InitialClaims'].map(lambda x: x.replace(',', ''))
Indiana_weekly_claims['InitialClaims'] = Indiana_weekly_claims['InitialClaims'].astype(int)
```

```python
#Iterate over the rows and add the data depending on the week of the year
for index, row in Indiana_weekly_claims.iterrows():
    Indiana_weekly_claims.loc[index, 'Date'] = allfridays(2020)[Indiana_weekly_claims.loc[index, 'Week']]
```

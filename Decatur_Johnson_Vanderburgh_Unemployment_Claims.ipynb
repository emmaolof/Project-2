```python
import pandas as pd
import csv
```

```python
def allfridays(year):
    return pd.date_range(start=str(year), end=str(year+1), 
                         freq='W-FRI').strftime('%Y%m%d').tolist()
```

```python
Vanderburgh_weekly_claims = pd.read_csv("vanderburgh_unemployment_claims.csv")
```

```python
Vanderburgh_weekly_claims.head(14)
```

```python
Vanderburgh_weekly_claims.columns = ['County','StateFips','CountyFips','Year', 'Month', 'MonthCode','Week', 'InitialClaims', 'ChangeFrom2019']
```

```python
Vanderburgh_weekly_claims['Date'] = allfridays(2020)[:14]
```

```python
Vanderburgh_weekly_claims['InitialClaims'] = Vanderburgh_weekly_claims['InitialClaims'].map(lambda x: x.replace(',', ''))
Vanderburgh_weekly_claims['InitialClaims'] = Vanderburgh_weekly_claims['InitialClaims'].astype(int)
```

```python
Decatur_weekly_claims = pd.read_csv("decatur_unemployment_claims.csv")
Decatur_weekly_claims.columns = ['County','StateFips','CountyFips','Year', 'Month', 'MonthCode','Week', 'InitialClaims', 'ChangeFrom2019']
Decatur_weekly_claims['Date'] = allfridays(2020)[:14]
```

```python
Johnson_weekly_claims = pd.read_csv("johnson_unemployment_claims.csv")
Johnson_weekly_claims.columns = ['County','StateFips','CountyFips','Year', 'Month', 'MonthCode','Week', 'InitialClaims', 'ChangeFrom2019']
Johnson_weekly_claims['Date'] = allfridays(2020)[:14]
Johnson_weekly_claims['InitialClaims'] = Johnson_weekly_claims['InitialClaims'].map(lambda x: x.replace(',', ''))
Johnson_weekly_claims['InitialClaims'] = Johnson_weekly_claims['InitialClaims'].astype(int)
```

```python
Decatur_Covid = pd.read_pickle('decatur_covid.pkl')
Johnson_Covid = pd.read_pickle('johnson_covid.pkl')
Vanderburgh_Covid = pd.read_pickle('vanderburgh_covid.pkl')
```

```python
Vanderburgh_Covid_Weekly_Claims = pd.merge(Vanderburgh_weekly_claims[['County', 'StateFips', 'CountyFips', 'InitialClaims', 'Date']], 
                                           Vanderburgh_Covid[['Count', 'Date']], how='outer', on='Date', sort=True)
```
